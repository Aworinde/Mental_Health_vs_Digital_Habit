```python
import seaborn as sns
```

```python
# Count frequency of each screen_time_hours value
mh2['Frequency'] = mh2.groupby('screen_time_hours')['screen_time_hours'].transform('count')
```

```python
plt.hist(mh2['screen_time_hours'], bins=20,color='#CC5500', edgecolor='white')
plt.title('Distribution of Screen Time')
plt.xlabel('Screen Time (Hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
mh2['Other_apps_hours'] = mh2['screen_time_hours'] - mh2['hours_on_TikTok']
```

```python
total_screen = mh2['screen_time_hours'].sum()
total_tiktok = mh2['hours_on_TikTok'].sum()
total_other = mh2['Other_apps_hours'].sum()

# Labels and values
labels = ['TikTok', 'Other Social Media Platforms']
sizes = [total_tiktok, total_other]
colors = ['#CC5500', '#6BA292']

# Plot doughnut chart
fig, ax = plt.subplots(figsize=(6, 6))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'width': 0.4, 'edgecolor': 'white'}
)

# Add center circle for doughnut look
centre_circle = plt.Circle((0, 0), 0.30, fc='white')
fig.gca().add_artist(centre_circle)

plt.title('Proportion of Screen Time: TikTok vs Other Social Media Platforms')
plt.tight_layout()
plt.show()

```

### Impact of screen time on stress level and mood score

```python
screen_time_hours = mh2['screen_time_hours']
stress_level = mh2['stress_level']
mood_score = mh2['mood_score']
sizes = mood_score * 10  # Bubble size multiplier

plt.figure(figsize=(8,6))
scatter = plt.scatter(screen_time_hours, stress_level, s=sizes, c=mood_score, cmap='viridis', alpha=0.6, edgecolors='w', linewidth=0.5)

plt.title('Screen Time vs Stress Level showing Mood Score')
plt.xlabel('Screen Time (Hours)')
plt.ylabel('Stress Level')

# Add a colorbar for mood score
cbar = plt.colorbar(scatter)
cbar.set_label('Mood Score')

plt.grid(True)
plt.show()
```

### Impact of screen time on sleep hours

```python
# Create screen time categories
# 0–3 hours → "Very Low", 3–6 hours → "Low", 6–9 hours → "Medium", 9–12 hours → "High"
mh2.loc[:, 'screen_time_category'] = pd.cut(
    mh2['screen_time_hours'],
    bins=[0, 3, 6, 9, 12, float('inf')],
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'],
)

plt.figure(figsize=(8,6))
sns.boxplot(
    x='screen_time_category',
    y='sleep_hours',
    data=mh2,
    color='#CC5500',  # Use a clear, colorful palette (like 'Set1', 'Set2', 'Pastel1', etc.)
    showmeans=True,
    meanprops={
        "marker": "o", 
        "markerfacecolor": "blue",  # Red mean point for contrast
        "markeredgecolor": "blue",
        "markersize": 6})

plt.title('Sleep Hours by Screen Time Category')
plt.xlabel('Screen Time Category')
plt.ylabel('Sleep Hours')
plt.grid(True)
plt.tight_layout()
plt.show()
```

It can be concluded that:
* Sleep hours across all groups generally range from 3 to 10 hours.
* Most users sleep between 6 and 8hours. There may be very short or very long sleepers, but the main distribution is fairly stable.
* Very Low, Low, and Medium screen time users have a median around 7 hours of sleep while High and Very High screen time users have a slightly lower median (~6.8–6.9 hours).

### Conclusion

The analysis shows that while screen time doesn't seem to impact sleep hours directly in this dataset, it does correlate with higher stress and lower mood, both of which in turn have strong correlations with sleep. In other words, screen time may be affecting sleep indirectly through stress. 

These insights underscore the importance of managing daily screen time to promote better sleep health. Further studies could explore underlying factors such as device usage before bedtime and the role of screen content to develop targeted recommendations for improving sleep hygiene.
